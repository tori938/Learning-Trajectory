```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
#for dark theme
plt.style.use('default')
```

```python
sns.__version__
```

```python
#export the file
covid_df = pd.read_csv('data/combined_covid_data.csv', sep=',')
covid_df.head(2)
```

create a cropped version of the file for further working out

```python
countries = ['Russia', 'Australia', 'Germany', 'Canada', 'United Kingdom']
# basically, filtering
croped_covid_df = covid_df[covid_df['country'].isin(countries)]

populations = pd.DataFrame([
    ['Canada', 37664517],
    ['Germany', 83721496],
    ['Russia', 145975300],
    ['Australia', 25726900],
    ['United Kingdom', 67802690]
    ],
    columns=['country', 'population']
)
croped_covid_df = croped_covid_df.merge(populations, on=['country'])
croped_covid_df['daily_confirmed_per_hundred'] = croped_covid_df['daily_confirmed'] / croped_covid_df['population'] * 100
croped_covid_df.head()
```

build two histograms

```python
#set the canvas
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

#first histogram
sns.histplot(
    data=croped_covid_df,
    x='daily_confirmed_per_hundred',
    bins=25,
    kde=True, #with a line
    ax=axes[0]
);
#set the title for the first histogram
axes[0].set_title('    100 ', fontsize=16)

#second histogram
sns.histplot(
    data=croped_covid_df,
    x='daily_confirmed_per_hundred',
    y='country',
    bins=25,
    color='red',
    ax=axes[1]
);
```

build a boxplot

```python
#set the canvas
fig = plt.figure(figsize=(10, 7))

boxplot = sns.boxplot(
    data=croped_covid_df,
    y='country',
    x='death_rate',
    orient='h',
    width=0.9
)
boxplot.set_title('   ');
boxplot.set_xlabel('');
boxplot.set_ylabel('');
boxplot.grid()
```

build a barchart

```python
#set the canvas
fig = plt.figure(figsize=(10, 7))

#convert to datetime
croped_covid_df['date'] = pd.to_datetime(croped_covid_df['date'])

#create the quarters
croped_covid_df['quarter'] = croped_covid_df['date'].dt.quarter

barplot = sns.barplot(
    data=croped_covid_df,
    x='country',
    y='daily_confirmed_per_hundred',
    hue='quarter',
)
barplot.set_title('     ');
```

build a jointplot: histogram + scatter graph

```python
jointplot = sns.jointplot(
    data=croped_covid_df, 
    x='people_fully_vaccinated_per_hundred', 
    y='daily_confirmed_per_hundred',
    hue='country',
    xlim = (0, 40),
    ylim = (0, 0.1),
    height=8,
)
```

create a heatmap

```python
pivot = croped_covid_df.pivot_table(
    values='people_vaccinated_per_hundred',
    columns='date',
    index='country',
)
#but convert columns to string from datetime
pivot.columns = pivot.columns.astype('string')
display(pivot)
```

```python
heatmap = sns.heatmap(data=pivot, cmap='YlGnBu')
heatmap.set_title('  ', fontsize=16);
```
