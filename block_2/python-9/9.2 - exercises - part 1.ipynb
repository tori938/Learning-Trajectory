EXERCISE 8.5: DISTANCE BETWEEN TWO VECTORS > 100

```python
a = np.array([23, 34, 27])
b = np.array([-54, 1, 46])
c = np.array([46, 68, 54])

d_ab = np.linalg.norm(a-b)
d_ac = np.linalg.norm(a-c)
d_bc = np.linalg.norm(b-c)

print(d_ab, d_ac, d_bc)
```

EXERCISE 8.6: ANY SCALAR PRODUCT PAIRS

```python
a = np.array([23, 34, 27])
b = np.array([-54, 1, 46])
c = np.array([46, 68, 54])

a_b = np.dot(a, b)
a_c = np.dot(a, c)
b_c = np.dot(b, c)

print(a_b, a_c, b_c)
print()

ab = np.sum(a*b)
ac = np.sum(a*c)
bc = np.sum(b*c)

print(ab, ac, bc)
```
