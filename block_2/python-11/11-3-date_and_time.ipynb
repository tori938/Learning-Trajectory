import libraries

```python
import pandas as pd, numpy as np
```

read csv property file

```python
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
melb_data.head()
```

create a copy of the property file

```python
melb_df = melb_data.copy()
melb_df.head()
```

convert date: from dd/mm/yyyy -> to standard datetime of yyyy/mm/dd

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
display(melb_df['Date'])
```

using accessory dt.? can find useful information \
.dt.year

```python
years_sold = melb_df['Date'].dt.year
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
```

which are the most popular sale months?

```python
melb_df['MonthSale'] = melb_df['Date'].dt.month
melb_df['MonthSale'].value_counts(normalize=True) * 100
```

or can use the mode() function to find the most popular month

```python
melb_df['MonthSale'] = melb_df['Date'].dt.month
melb_df['MonthSale'].mode()
```

working with time intervals

```python
delta_days = melb_df['Date'] - pd.to_datetime('2016-01-01') 
display(delta_days.dt.days)
```

age of the property: from the day it's build, to the day it's sold

```python
melb_df['BuildingAge'] = melb_df['Date'].dt.year - melb_df['YearBuilt']
display(melb_df['BuildingAge'])
```

don't duplicate data: BuildingAge > YearBuilt

```python
melb_df = melb_df.drop('YearBuilt', axis=1)
melb_df
```

Exercise 3.5: find the average time interval between each observation in the state of Nevada in days

```python
#let's convert Time to all

#create a new column - Month
usa_et_data['Month'] = usa_et_data['Time'].dt.month

#create a new column - Day
usa_et_data['Day'] = usa_et_data['Time'].dt.day

#create a new column - Date
usa_et_data['Date'] = usa_et_data['Time'].dt.date

usa_et_data.head()
```

find the data on Nevada only

```python
nv_only = usa_et_data[usa_et_data['State'] == 'NV']
nv_only.head()
```

find the difference between each observation in days, \
and then find the average

```python
dif_date = nv_only['Date'].diff()
dif_date_in_days = dif_date.dt.days
avg_maybe = dif_date_in_days.mean()
print(round(avg_maybe, 0))
```
