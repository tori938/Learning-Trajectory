EXERCISE 2.3

```python
customer_df = pd.DataFrame({
            'number': [0, 1, 2, 3, 4],
            'cust_id': [128, 1201, 9832, 4392, 7472],
            'cust_age': [13, 21, 19, 21, 60],
            'cust_sale': [0, 0, 0.2, 0.15, 0.3],
            'cust_year_birth': [2008, 2000, 2002, 2000, 1961],
            'cust_order': [1400, 14142, 900, 1240, 8430]
        })

customer_df2 = pd.DataFrame({
            'number2': [0, 1, 2, 3, 4],
            'cust_id2': [128, 1201, 9832, 4392, 7472],
            'cust_age2': [13, 21, 19, 21, 60],
            'cust_sale2': [0, 0, 0.2, 0.15, 0.3],
            'cust_year_birth2': [2008, 2000, 2002, 2000, 1961],
            'cust_order2': [1400, 14142, 900, 1240, 8430]
        })

def delete_columns(df, col=[]):
    #work with a copy of the dataframe directly -> highly important otherwise leads to errors
    new_df = df
    #find whether the name of the column exists in the list
    for column in col:
        #if the name exists in the column names
        if column in df.columns: #-> no need to use the list()
            #delete the column 
            new_df = new_df.drop(column, axis=1, errors='ignore') #-> work directly with new_df entire time
        else:
            #if name of the column doesn't exist
            return None
    return new_df

print('first time')
print(delete_columns(df = customer_df, col = ['number']))
print()

print('second time')
print(delete_columns(col = ['cust_sale2', 'cust_age2'], df = customer_df2))
print()

print('third time')
print(delete_columns(df = customer_df2, col = ['cust_sale2', 'non-existent']))
print()
```

EXERCISE 2.4

```python
countries_df = pd.DataFrame({
    'Country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'Population (m)': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'Squared Area': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
countries_df
```

Для каждой страны рассчитайте плотность населения (количество человек на квадратный километр).

```python
countries_copy = countries_df.copy()
#create a new column with population in millions
countries_copy['Population'] = countries_copy['Population (m)'] * 1000000
#create a new column with population density
countries_copy['Population Density'] = countries_copy['Population'] / countries_copy['Squared Area']
#delete the column - population (m) that I no longer need
countries_copy = countries_copy.drop('Population (m)', axis=1)
countries_copy
```

Затем по полученным данным рассчитайте среднее по плотностям населения в указанных странах. Ответ округлите до сотых.

```python
avg = countries_copy['Population Density'].mean()
print(avg, round(avg, 2), sep='\n')
```
