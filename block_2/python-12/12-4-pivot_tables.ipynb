```python
import pandas as pd, numpy as np
```

export the new updated property file

```python
melb_data = pd.read_csv('data/melb_data_fe_updated.csv', sep=',')
melb_data.head(2)
```

create a copy

```python
melb_df = melb_data.copy()
melb_df.head(2)
```

foundation: datetime and category

```python
#convert to datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

#convert to category
#do not include specific columns
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
#set the max no of unique categories
max_unique_count = 150
#loop through column names
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: #check the condition
        melb_df[col] = melb_df[col].astype('category') #change the column type
display(melb_df.info())
```

```python
melb_df.groupby('Rooms')[['Price', 'BuildingArea']].median()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean()
```

to convert to a pivot table alike

```python
round(melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack(), 2)
```

convert to a pivot table

```python
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

price difference by weekend for every region for no. of properties

```python
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
)
```

median and mean difference by landsize for each region and type of property

```python
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

multidimensional pivot tables

```python
melb_df.pivot_table(
    values='Price',
    index=['Method','Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

accessing data in the pivot table: \
input the prior test data into a variable

```python
pivot = melb_df.pivot_table(
    values ='Landsize',
    index ='Regionname',
    columns ='Type',
    aggfunc =['median', 'mean'],
    fill_value=0
)
```

find variable = pivot 's columns

```python
pivot.columns
```

thus can easily find the average landsize of each unit in each region

```python
display(pivot['mean']['unit'])
```

can filter data: \
regions with average landsize is smaller than their median landsize

```python
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
display(filtered_pivot)
```

only returning the regions' index

```python
print(list(filtered_pivot.index))
```
