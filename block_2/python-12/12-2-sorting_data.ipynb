```python
import pandas as pd, numpy as np
```

export the new updated property file

```python
melb_data = pd.read_csv('data/melb_data_fe_updated.csv', sep=',')
melb_data.head(2)
```

create a copy

```python
melb_df = melb_data.copy()
melb_df.head(2)
```

foundation: datetime and category

```python
#convert to datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

#convert to category
#do not include specific columns
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
#set the max no of unique categories
max_unique_count = 150
#loop through column names
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: #check the condition
        melb_df[col] = melb_df[col].astype('category') #change the column type
display(melb_df.info())
```

sort the data by ASC price

```python
melb_df.sort_values(by='Price').head(3)
```

sort the data by DESC date

```python
melb_df.sort_values(by='Date', ascending=False).head(3)
```

sort by ASC distance + price

```python
melb_df.sort_values(by=['Distance', 'Price']).head(2)
```

can use index to show only specific columns

```python
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

combining sorting with filtering

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```
