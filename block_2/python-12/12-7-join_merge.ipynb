```python
import pandas as pd, numpy as np
```

export the movies data files

```python
movies_data = pd.read_csv('data/movies_data/movies.csv', sep=',')
dates_data = pd.read_csv('data/movies_data/dates.csv', sep=',')
rate1_data = pd.read_csv('data/movies_data/ratings1.csv', sep=',')
rate2_data = pd.read_csv('data/movies_data/ratings2.csv', sep=',')
```

```python
#join two ratings tables
ratings_data = pd.concat(
    [rate1_data, rate2_data],
    ignore_index=True
)

#delete any duplicated
ratings_data = ratings_data.drop_duplicates(ignore_index=True)

#join ratings and dates tables
ratings_dates_data = pd.concat([ratings_data, dates_data], axis=1)
```

```python
joined_false = ratings_dates_data.join(
    movies_data,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

looks to be an sql inner join equivalent

```python
joined = ratings_dates_data.join(
    movies_data.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
```

merge() method works with keys directly

```python
merged = ratings_dates_data.merge(
    movies_data,
    on='movieId',
    how='left'
)
display(merged.head())
```

check that the no. of rows are equal

```python
print('no. of rows in ratings_dates: ', ratings_dates_data.shape[0])
print('no. of rows in merged: ', merged.shape[0])
print(ratings_dates_data.shape[0] == merged.shape[0])
```

using outer() leaves with extra records that do not yet have a rating

```python
merged2 = ratings_dates_data.merge(
    movies_data,
    on='movieId',
    how='outer'
)
print('no. of rows in merged2: ', merged2.shape[0])
display(merged2.tail())
```

using merge() method to combine the ratings tables

```python
merge_ratings = rate1_data.merge(rate2_data, how='outer')
print('no. of rows in merge_ratings: ', merge_ratings.shape[0])
display(merge_ratings)
```

```python
items_df = pd.DataFrame({
            'item_id': [417283, 849734, 132223, 573943, 19475, 3294095, 382043, 302948, 100132, 312394],
            'vendor': ['Samsung', 'LG', 'Apple', 'Apple', 'LG', 'Apple', 'Samsung', 'Samsung', 'LG', 'ZTE'],
            'stock_count': [54, 33, 122, 18, 102, 43, 77, 143, 60, 19]
        })
display(items_df)
```

```python
purchase_df = pd.DataFrame({
            'purchase_id': [101, 101, 101, 112, 121, 145, 145, 145, 145, 221],
            'item_id': [417283, 849734, 132223, 573943, 19475, 3294095, 382043, 302948, 103845, 100132],
            'price': [13900, 5330, 38200, 49990, 9890, 33000, 67500, 34500, 89900, 11400]
        })
display(purchase_df)
```

create merged \
so that only existing items are combined with available stock

```python
merged_i = items_df.merge(
    purchase_df,
    on='item_id',
    how='inner'
)
display(merged_i)
```

find the total amount of sales

```python
income = merged_i['stock_count'] * merged_i['price']
print(income)
print()

print(income.sum())
```
