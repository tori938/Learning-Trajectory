extract the year

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

Exercise 8.1: \

```python
movies_ratings['year_release'] = movies_ratings['title'].apply(get_year_release)
```

Exercise 8.2: which film released in 1999 got the lowest average rating?

```python
#filter the data set by release year
release_year_1999 = movies_ratings[movies_ratings['year_release'] == 1999]

#group by the title, and find the lowest average rating by using sorting ASC
release_year_1999.groupby('title')['rating'].mean().sort_values(ascending=True)
```

Exercise 8.3: which films released in 2010 got the lowest average reting by genre?

```python
#filter the data set by release year
release_year_2010 = movies_ratings[movies_ratings['year_release'] == 2010]

#group by genre, and find the lowest average rating by sorting ASC
release_year_2010.groupby('genres')['rating'].mean().sort_values(ascending=True)
```

Exercise 8.4: find the user with the highest no. of unique genre combinations

```python
#group by user, and find descriptive statistics for genres
#sorting by agg: unique in DESC order
movies_ratings.groupby('userId')['genres'].agg('describe').sort_values(by='unique', ascending=False)
```

Exercise 8.5: find the user with the lowest no. of ratings made, yet the highest average film rating

```python
#group by user, and find agg: count and mean
#sorting by agg: count in ASC and mean in DESC
movies_ratings.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[True, False])
```

Exercise 8.6: find the genres combination for 2018 with the highest average rating with the no. of ratings made > 10

```python
#filter the data set by release year
release_year_2018 = movies_ratings[movies_ratings['year_release'] == 2018]

#group by genre, and find the average rating and no. of ratings made
#sorting by agg: mean in DESC
high_avg_cnt = release_year_2018.groupby('genres')['rating'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)

#filter the created dataframe by highest count possible
high_avg_cnt[high_avg_cnt['count'] > 10]
```

Exercise 8.7: \
create a new feature year_rating
create a pivot table: values = rating, index = year_rating, columns = genre, aggfunc = mean

```python
#convert date to datetime
movies_ratings['date'] = pd.to_datetime(movies_ratings['date'])

#create a new feature: year_rating
movies_ratings['year_rating'] = movies_ratings['date'].dt.year
```

```python
movies_ratings.head(3)
```

delete unnecessary columns, they are bothering me

```python
movies_ratings = movies_ratings.drop(['Unnamed: 0', 'date'], axis=1, errors='ignore')
```

```python
pt_year_genre = movies_ratings.pivot_table(
    values ='rating',
    index = ['year_rating'],
    columns ='genres',
    aggfunc ='mean',
    fill_value = 0
)
display(pt_year_genre)
```

```python
display(pt_year_genre['Action|Adventure|Animation|Children|Comedy|IMAX'])
```
