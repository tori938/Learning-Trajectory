Exercise 2.4: build a boxplot showing the link between the flat price and region ecology

```python
sns.boxplot(
    data = sber_data,
    x='price_doc',
    y='ecology',
    width = 0.7
    );
```

Exercise 2.5: build a scatter plot showing how the flat price is linked to the distance from the center of the city.

```python
fig = plt.figure(figsize = (6, 4))
axes = fig.add_axes([0, 0, 1, 1])

axes.scatter(
    x=sber_data['price_doc'],
    y=sber_data['kremlin_km'],
    s = 5,
    marker = 'x',
    c = 'blue'
)
axes.set_xlabel('price')
axes.set_ylabel('distance')
axes.grid()
axes.set_title('link between price and distance from city centre', fontsize=14);
```

apply isnull() method on columns

```python
cols_null_percent = sber_data.isnull().mean() * 100 #convers to %
cols_with_null = cols_null_percent[cols_null_percent>0].sort_values(ascending=False)
display(cols_with_null)
```

```python
type(cols_with_null)
```

```python
cols_with_null.index
```

build a bar chart to short columns with null values

```python
cols_with_null.plot(
    kind='bar',
    figsize=(10, 5),
    title='Распределение пропусков в данных'
);
```

using a heatmap to show columns with null values

```python
colors = ['blue', 'yellow'] 
fig = plt.figure(figsize=(10, 4))
cols = cols_with_null.index
ax = sns.heatmap(
    sber_data[cols].isnull(),
    cmap=sns.color_palette(colors),
)
```
